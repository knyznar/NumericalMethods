```python
#ZADANIE 8
import numpy as np

tab = [1, -7, 3, 43, -28, -60]
print(np.roots(tab))

def check(x):
    w = x**5 - 7*x**4 + 3*x**3 + 43*x**2 - 28*x - 60  
    if (w != 0):
        print(x, " nie jest pierwiastkiem wielomianu")
    else:
        print(x, " jest pierwiastkiem wielomianu")
        
check(5)
check(1)
```

```python
#ZADANIE 10
import matplotlib.pyplot as plt

def funkcja(x):
    return x**3 - 3*x

x = np.linspace(-1,1)
y = [funkcja(i) for i in x]
plt.plot(x, y)
plt.title('Zakres <-1,1>')
plt.xlabel('x')
plt.ylabel('y')
plt.legend('Wykres 1')
plt.show()

x = np.linspace(-5,5)
y = [funkcja(i) for i in x]
plt.plot(x, y)
plt.title('Zakres <-5,5>')
plt.xlabel('x')
plt.ylabel('y')
plt.legend('Wykres 2')
plt.show()

x = np.linspace(0,5)
y = [funkcja(i) for i in x]
plt.plot(x, y)
plt.title('Zakres <0,5>')
plt.xlabel('x')
plt.ylabel('y')
plt.legend('Wykres 3')
plt.show()

```
